```python

import pandas as pd
import numpy
import os
import seaborn
import matplotlib.pyplot as plt

data = pd.read_csv(r'C:\Users\tikun\Documents\Data analysis projects\Alcohol Consumption\nesarc_pds.csv')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)


pd.set_option('display.float_format', lambda x:'%f'%x)


data['S2AQ8A'] = pd.to_numeric(data['S2AQ8A'],errors='coerce')
data['S2AQ8B'] = pd.to_numeric(data['S2AQ8B'],errors='coerce')
data['S1Q12B'] = pd.to_numeric(data['S1Q12B'],errors='coerce')
data['S1Q11B'] = pd.to_numeric(data['S1Q11B'],errors='coerce')
data['S1Q10B'] = pd.to_numeric(data['S1Q10B'],errors='coerce')
data['S1Q6A'] = pd.to_numeric(data['S1Q6A'],errors='coerce')
data['AGE'] = pd.to_numeric(data['AGE'],errors='coerce')
```

```python

print("count of Total Household income in last 12 months:")
house=data["S1Q12B"].value_counts(sort=False)
print(house)
print("count of Total family income in last 12 months:")
fam=data["S1Q11B"].value_counts(sort=False)
print(fam)
print("count of Total personal income in last 12 months:")
person=data["S1Q10B"].value_counts(sort=False)
print(person)
print("count of : How often drank alcohol in last 12 months:")
freq=data["S2AQ8A"].value_counts(sort=False)
print(freq)
print("count of How often drank largest number of drink in last 12 months:")
large=data["S2AQ8D"].value_counts(sort=False)
print(large)
print("count of How often drank enough to feel intoxicated:")
intox=data["S2AQ10"].value_counts(sort=False)
print(intox)
print("count of Highest grade or year of school completed:")
grade=data["S1Q6A"].value_counts(sort=False)
print(grade)
```

```python

sub1=data[(data['AGE']>=18)]

sub2 = sub1.copy()


print("count of Total Household income in last 12 months:")
house=sub2["S1Q12B"].value_counts(sort=False)
print(house)
print("percentage of total household income in last 12 months:")
housep=sub2["S1Q12B"].value_counts(sort=False,normalize=True)
print(housep)


print("count of Total family income in last 12 months:")
fam=sub2["S1Q11B"].value_counts(sort=False)
print(fam)
print("percentage of total family income in last 12 months:")
famp=sub2["S1Q11B"].value_counts(sort=False,normalize=True)
print(famp)

print("count of Total personal income in last 12 months:")
person=sub2["S1Q10B"].value_counts(sort=False)
print(person)
print("percentage of total personal income in last 12 months:")
personp=sub2["S1Q10B"].value_counts(sort=False,normalize=True)
print(personp)

print("count of How often drank alcohol in last 12 months:")
freq=sub2["S2AQ8A"].value_counts(sort=False)
print(freq)
print("percentage of how often drank alcohol in last 12 months:")
freqp=sub2["S2AQ8A"].value_counts(sort=False,normalize=True)
print(freqp)

print("count of number of alcohols usually consumed when any kind of alcohol is consumed:")
num=sub2["S2AQ8B"].value_counts(sort=False)
print(num)
numq=sub2["S2AQ8B"].value_counts(sort=False,normalize=True)

print("count of How often drank largest number of drink in last 12 months:")
large=sub2["S2AQ8D"].value_counts(sort=False)
print(large)
print("percentage of how often drank largest number of drink in last 12 months:")
largep=sub2["S2AQ8D"].value_counts(sort=False,normalize=True)
print(largep)

print("count of How often drank enough to feel intoxicated:")
intox=sub2["S2AQ10"].value_counts(sort=False)
print(intox)
print("percentage of how often drank enough to feel intoxicated:")
intoxp=sub2["S2AQ10"].value_counts(sort=False,normalize=True)
print(intoxp)

print("count of Highest grade or year of school completed:")
grade=sub2["S1Q6A"].value_counts(sort=False)
print(grade)
print("percentage of highest grade or year of school completion:")
gradep=sub2["S1Q6A"].value_counts(sort=False,normalize=True)
print(gradep)

```

```python




sub2['S2AQ8A']=sub2['S2AQ8A'].replace(99, numpy.nan)
sub2['S2AQ8B']=sub2['S2AQ8B'].replace(99, numpy.nan)


sub2['Housecat']=pd.qcut(sub2.S1Q12B,3,labels=["Low","Medium","High"])
print("Categorized count of total household incomes:")
Hincat=sub2['Housecat'].value_counts(sort=False)
print(Hincat)
print("Categorized percentage of total household incomes:")
Hincatp=sub2['Housecat'].value_counts(sort=False,normalize=True)
print(Hincatp)

sub2['Famcat']=pd.qcut(sub2.S1Q11B,3,labels=["Low","Medium","High"])
print("Categorized count of total family income:")
Fincat=sub2['Famcat'].value_counts(sort=False)
print(Fincat)
print("Categorized percentage of total family income:")
Fincatp=sub2['Famcat'].value_counts(sort=False,normalize=True)
print(Fincatp)

sub2['Percat']=pd.qcut(sub2.S1Q10B,3,labels=["Low","Medium","High"])
print("Categorized count of total personal income:")
Pincat=sub2['Percat'].value_counts(sort=False)
print(Pincat)
print("Categorized percentage of total personal income:")
Pincatp=sub2['Percat'].value_counts(sort=False,normalize=True)
print(Pincatp)

recode1={1:365,2:360,3:208,4:104,5:52,6:36,7:12,8:11,9:6,10:2}
sub2['USFREQMO']=sub2['S2AQ8A'].map(recode1)
sub2['USFREQMO']= pd.to_numeric(sub2['USFREQMO'],errors='coerce')
user_freq=sub2['USFREQMO'].value_counts(sort=False)
print(user_freq)

sub2['Toxic_Lvl']=sub2['USFREQMO']*sub2['S2AQ8B']
sub2['Toxic_Lvl']= pd.to_numeric(sub2['Toxic_Lvl'],errors='coerce')
print("Intake rate count:")
toxiclvl=sub2['Toxic_Lvl'].value_counts(sort=False)
print(toxiclvl)
print("Intake rate percentage:")
toxiclvlp=sub2['Toxic_Lvl'].value_counts(sort=False,normalize=True)
print(toxiclvlp)




```

```python

sub2["Percat"] = sub2["Percat"].astype('category')

seaborn.countplot(x="Percat", data=sub2)
plt.xlabel('Different income groups')
plt.title('Alcohol consumption in different total personal income groups in the NESARC Study')

```

```python
sub2['Gradecat'] = pd.cut(sub2.S1Q6A, [0, 2,6,9,14])
sub2["Gradecat"] = sub2["Gradecat"].astype('category')

seaborn.countplot(x="Gradecat", data=sub2)
plt.xlabel('Highest level of education')
plt.title('Education status')

```

```python
seaborn.distplot(sub2["Toxic_Lvl"].dropna(), kde=False)
plt.xlabel('Amount of alcohol consumed per year')
plt.title('Estimated amount of alcohol consumed per Year among Young Adults in the NESARC Study')
```
